Data Cleaning

```python
# Calculate the number of null values in each column
null_values = ((app.isnull().sum()/len(app))*100).sort_values(ascending=False)

print(null_values)
```

```python
#Identify columns with more than 40% null values
columns_to_drop = null_values[null_values > 40].index
print("Number of Columns dropped: ", columns_to_drop.size)
for column in columns_to_drop:
  print(column)
```

```python
#drop columns
app_cleaned = app.drop(columns=columns_to_drop)
```
